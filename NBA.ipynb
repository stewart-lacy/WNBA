```python
file_to_load = "NOP.csv"
NOP = pd.read_csv(file_to_load)
pd.set_option('display.max_columns',None)
NOP_clean=NOP.drop(NOP.index[[13]])
NOP_clean
```

# 2 Point Shooting 

```python
two_point = NOP_clean[['Player','2P','2PA','2P%']]
two_point_perc=two_point.sort_values(by='2P%', ascending=False)
Format_dict_2ps = {'2P%':'{:.2%}'}
Two_Point_Shooting_perc=two_point_perc.style.format(Format_dict_2ps)
Two_Point_Shooting_perc
```

# Player that shoots the most Field Goals 

```python
Two_Point_FG=two_point.sort_values(by='2PA', ascending=False)
format_dict_2ps_perc = {'2P%':'{:.2%}'}
Two_Point_FGs = Two_Point_FG.style.format(format_dict_2ps_perc)
Two_Point_FGs
```

# 3 Point Shooting 

```python
three_point = NOP_clean[['Player','3P','3PA','3P%']]
three_point_drop_nan = three_point[three_point['3P%'].notna()]
Three_Point_Shooting_Perc=three_point_drop_nan.sort_values(by=['3P%'], ascending=False)
Three_Point_Shooting_Perc
format_dict = {'3P%': '{:.2%}'}
Three_Point_Shooting_Perctange=Three_Point_Shooting_Perc.style.format(format_dict)
Three_Point_Shooting_Perctange
```
